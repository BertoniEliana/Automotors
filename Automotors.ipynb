```python
#Eliminación de valores nulos: Las ultimas 921320 filas no contienen ningun dato
ventas_df.dropna(inplace=True)
```

```python
#Correccion de tipos de datos
columns_to_string = ['IdCliente', 'NombreCliente', 'Referencia', 'Familia', 'Localidad', 'Sede']
columns_to_int = ['Empleado', 'CodigoFamilia', 'Area']
ventas_df[columns_to_string] =  ventas_df[columns_to_string].astype('string')
ventas_df[columns_to_int] = ventas_df[columns_to_int].astype('int')
```

```python
ventas_df["Ventas"] = pd.to_numeric(ventas_df["Ventas"].str.replace(",", "").str.strip(), errors="coerce")
```

```python
ventas_df
```

```python
mediana_ventas = ventas_df["Ventas"].median()
```

```python
ventas_df["Ventas"].fillna(mediana_ventas)
```

```python
ventas_df["Fecha"] = pd.to_datetime(ventas_df["Fecha"], format="%d/%m/%Y")
```

```python
ventas_df
```

```python
duplicados = ventas_df[ventas_df.duplicated(subset=['IdCliente', 'Fecha', 'Empleado', 'Referencia', 'Cantidad'], keep='first')]
```

```python
duplicados
```

```python
ventas_df = ventas_df.drop_duplicates(subset=['IdCliente', 'Fecha', 'Empleado', 'Referencia', 'Cantidad'], keep='first')
```

```python
ventas_df.dtypes
```

## Analisis Exploratorio de Datos

```python
ventas_y_empleados = empleados_df.merge(ventas_df, left_on='Id_Empleado', right_on='Empleado', how='left')
```

```python
ventas_y_empleados.drop(columns=['Empleado'], inplace=True)
```

```python
ventas_y_empleados.rename(columns={"Nombre y Apellido": "NombreEmpleado"}, inplace=True)
```

```python
ventas_y_empleados.head(5)
```

```python
ventas_y_empleados['VentasUSD'] = (ventas_y_empleados['Ventas'] * 0.00023).round(2)
```

```python
ventas_y_empleados = ventas_y_empleados.drop(columns = ['Ventas'])
```
